## 계절 성분 처리
- This is a naive decomposition. More sophisticated methods should be preferred.<br>
The additive model is Y[t] = T[t] + S[t] + e[t]<br>
The multiplicative model is Y[t] = T[t] * S[t] * e[t]

- T(trend) = 추세 성분, S(sesonal) = 계절 성분, e 혹은 R(residual) = 불규칙 성분 <br>
시계열은 이렇게 세 가지 성분으로 분해할 수 있습니다.

- additive와 multiplicative 차이 <br>
1. additive에서 trend와 frequency가 다릅니다. 장기간을 고려한 분석에선 <br>
sesonal 결과가 trend를 따라가지 못해 적합하지 않습니다.
2. 반면, multi 모델은 분해값들의 곱인 비율로 이루어져 있어서 트렌드에 맞게 sesonal 데이터가 변화합니다. <br>
시간이 지남에 따라 금리가 오르는 현상을 고려할 수 있는 모델로 볼 수 있습니다.

**따라서, 시간의 흐름에 따라 계절성이 변하지 않는다면 additive, <br>
데이터의 양에 따라 데이터 패턴이 달라지는 경우엔 multiplicative 모델을 사용합니다.**

```python
## Decomposition
decomposition = sm.tsa.seasonal_decompose(data['Passengers'],
                                          model='multiplicative')
fig = decomposition.plot()
fig.set_size_inches(10, 10)
plt.show()
```

```python
decomposition.trend.dropna()
```

# ACF, PACF TEST(자기상관함수)
- 정상계열은 빠르게 0으로 내려갑니다.<br>
  약한 정상의 경우 분산이 시차에 따라 변하기 때문에
  
- 비정상 계열의 ACF는 천천히 감소하며, 큰 값을 가지는 경우가 많습니다.


```python
# split => 시계열에서는 다르게 => shuffle=False
train_data, test_data = train_test_split(data, test_size=0.2, shuffle=False)
```

```python
# 빠르게 값을 내려가야 정상
# 그렇지 않다면, 잘 안되고 있는 것
fig, ax = plt.subplots(1,2, figsize=(10,5))
fig.suptitle('Data')

sm.graphics.tsa.plot_acf(train_data.values.squeeze(), lags=50, ax=ax[0])
sm.graphics.tsa.plot_pacf(train_data.values.squeeze(), lags=50, ax=ax[1])
```

```python
import numpy as np
log_train_data  = np.log(train_data.copy())
log_test_data   = np.log(test_data.copy())
diff_train_data = log_train_data['Passengers'].diff() # 차분 수행 (xt - xt-1)
diff_train_data.dropna(inplace=True)

diff_train_data
```

```python
diff_train_data.plot()
```

```python
fig, ax = plt.subplots(1,2, figsize=(10,5))
fig.suptitle('Diff Data')

sm.graphics.tsa.plot_acf(diff_train_data.values.squeeze(), lags=30, ax=ax[0])
sm.graphics.tsa.plot_pacf(diff_train_data.values.squeeze(), lags=30, ax=ax[1])
```

```python
                                            # p, d, q
model = SARIMAX(log_train_data.values, order=(1, 0, 0)
#, seasonal_order=(2,1,1,12)
) # AR(1)
history = model.fit()
history.summary()
```

## 최적의 p, d, q 값 찾기

```python
p = range(0, 3) # 0,1,2
d = range(0, 2) # 0,1
q = range(0, 3) # 0,1,2

pdq = list(itertools.product(p, d, q))

aic = []

for i in pdq:
  model = SARIMAX(log_train_data.values, order=(i))
  model_fit = model.fit()
  aic.append(model_fit.aic)
```

```python
aic
```

```python
optimal = [(pdq[i], j) for i, j in enumerate(aic) if j == min(aic)]
optimal
```

```python
opt_model = SARIMAX(log_train_data.values, order=optimal[0][0])
opt_model_fit = opt_model.fit()

opt_model_fit.summary()
```

```python
from statsmodels.tsa.stattools import adfuller
```

```python
result = adfuller(train_data)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
result = adfuller(np.log(train_data))
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```

```python
result = adfuller(diff_train_data)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))
```
